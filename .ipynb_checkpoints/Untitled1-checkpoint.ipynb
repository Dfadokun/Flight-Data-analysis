##### A very importance question comes to our mind. What are the the reasons for this cancellations?

From the analysis of 2019 flight data, one key reason stands out Weather. Over, 45000 flights were cancelled due to unfavourable weather condition. the true is, can we do anything about this. Certainly NO!!!. We just have to live with severe thhunderstoms, poor visibility, windstorms, hurricanes etc. We came to meet them and they will surely outlive us.

Weather is untouchable, next we look at carrier cancellations which are majorly caused by airlines. Over 30% of the cancellation were as a result of inefficient management of operations by airlines in 2019. Can we do something better

NAS and Security reason are majorly from airport and air traffic control managment. Nobody will compain if there are cancellations due to security issues. Imagine if a terror attack is imminent, nobody will board a plane. 


```python


Weather_Cancel = []
NAS_Cancel = []
Carrier_Cancel=[]
Security_Cancel=[]

for i in air_time2['CANCELLATION_CODE']:
    if i=='A':
        Carrier_Cancel.append(i)
    else:
        Carrier_Cancel.append(np.nan)
air_time2.loc[:,'Carrier_Cancel']=Carrier_Cancel


for i in air_time2['CANCELLATION_CODE']:
    if i=='B':
        Weather_Cancel.append(i)
    else:
        Weather_Cancel.append(np.nan)
air_time2.loc[:,'Weather_Cancel']=Weather_Cancel       


for i in air_time2['CANCELLATION_CODE']:
    if i=='C':
        NAS_Cancel.append(i)
    else:
        NAS_Cancel.append(np.nan)
air_time2.loc[:,'NAS_Cancel']=NAS_Cancel


for i in air_time2['CANCELLATION_CODE']:
    if i=='D':
        Security_Cancel.append(i)
    else:
        Security_Cancel.append(np.nan)
air_time2.loc[:,'Security_Cancel']=Security_Cancel
```

```python
yz=air_time2.groupby(['AIRLINE'],as_index=False)[['Carrier_Cancel','Weather_Cancel', 'NAS_Cancel','Security_Cancel']].count()
yz.tail()
```

```python
fig,ax=plt.subplots(figsize=(12,10))
w=0.40

ind1=np.arange(0,len(yz['AIRLINE'])*2,2)
#ind2 = [i + w for i in ind1]
#ind3 = [i + w for i in ind2]
#ind4 = [i + w for i in ind3]
plt.barh(ind1-w/2,yz['Carrier_Cancel'],w,label="Carrier_Cancel")
plt.barh(ind1+w/2,yz['Weather_Cancel'],w,label="Weather_Cancel")
plt.barh((ind1+w/2+w),yz['NAS_Cancel'],w,label="NAS_Cancel")
plt.barh((ind1+w/2+2*w),yz['Security_Cancel'],w,label="Security_Cancel")
plt.yticks([0, 2, 4, 6, 8],['American Airlines Inc.', 'Delta Air Lines Inc.', 'Skywest Airlines Inc.', 'Southwest Airlines Co.', 'United Air Lines Inc.'],rotation=0, fontsize=12)
plt.legend()
plt.ylabel('Airlines',fontsize=15)
plt.xlabel("Cancellation Count(s)",fontsize=15)
plt.xlim(0,20000)
plt.title('Flight Cancellations Reasons for Top Five Airline In 2019', fontsize=18)
for p in ax.patches:
    percent ='{:,.0f}'.format(p.get_width())
    width, height =p.get_width(),p.get_height()
    x=p.get_x()+width+0.02
    y=p.get_y()+height/2
    ax.annotate(percent,(x,y), fontsize=12)
```

```python
delay_sum=air_time.loc[:,['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY',
       'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']].sum().nlargest(5)
idx=air_time.loc[:,['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY',
       'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']].sum().nlargest(5).index
delay_count=air_time.loc[:,['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY',
       'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']].count(numeric_only=True).nlargest(5)
idx2=air_time.loc[:,['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY',
       'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']].count(numeric_only=True).nlargest(5).index
```

```python
fig,ax=plt.subplots(figsize=(10,8))
plt.barh(idx2,delay_count)
plt.ylabel("Delay Reasons",fontsize=15)
plt.xlabel("Delay counts",fontsize=15)
plt.xlim(0,1000000)
plt.title('Reason For Flight Delays For Top Five Airlines In 2019', fontsize=18)

for p in ax.patches:
    percent ='({:,.0f})'.format(p.get_width())
    width, height =p.get_width(),p.get_height()
    x=p.get_x()+width+0.02
    y=p.get_y()+height/1.5
    ax.annotate(percent,(x,y),fontsize=12)
total = delay_count.sum()   
for p in ax.patches:
    percent ='({:,.2f})'.format(100*p.get_width()/total)
    width, height =p.get_width(),p.get_height()
    x=p.get_x()+width+0.02
    y=p.get_y()+height/1.1
    ax.annotate(percent,(x,y),fontsize=12)
```

```python
INDEX=Airline_mins.loc[:,['AIRLINE','CARRIER_DELAY']]

INDEX['DELAY_INDEX']=(INDEX['CARRIER_DELAY']/INDEX['CARRIER_DELAY'].sum())
INDEX.sort_values(by="DELAY_INDEX",inplace=True, ascending=False)
```

To conclude a matrix was developed to rank airline depict how each airline performed interms of delay and cancellation. This matrix is based strictly on Carrier delays or cancellation as this are solely from the airlines. The lower the values the beeteer the performance.

```python
fig,ax=plt.subplots(figsize=(10,8))
plt.barh(INDEX['AIRLINE'],INDEX['DELAY_INDEX'])
plt.ylabel("Airline",fontsize=15)
plt.xlabel("Delay Index",fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Delay Matrix For To Five Airlines in 2019', fontsize=18)

#plt.xlim(0,0.30)

for p in ax.patches:
    percent ='({:,.2f})'.format(p.get_width())
    width, height =p.get_width(),p.get_height()
    x=p.get_x()+width+0.00001
    y=p.get_y()+height/2
    ax.annotate(percent,(x,y),fontsize=12)
```

The Delays matrix shows that united airline is the best performing in this area but we need to bear in mind that United is actually the least active of the top 5 airlines. Delta airline also showed a very good performance considering it second most active airline.

```python
INDEX1=yz.loc[:,['AIRLINE','Carrier_Cancel']]

INDEX1['Carrier_Cancel'].sum()

INDEX1['CANCELLATION_INDEX']=(INDEX1['Carrier_Cancel']/INDEX1['Carrier_Cancel'].sum())
INDEX1.sort_values(by="CANCELLATION_INDEX",inplace=True,ascending=False)
```

```python
fig,ax=plt.subplots(figsize=(10,8))
plt.barh(INDEX1['AIRLINE'],INDEX1["CANCELLATION_INDEX"])
plt.ylabel("Airline",fontsize=15)
plt.xlabel("Cancellation Index",fontsize=15)
#plt.xlim(0,0.30)
plt.title('Cancellation Matrix For To Five Airlines In 2019', fontsize=18)
for p in ax.patches:
    percent ='({:,.2f})'.format(p.get_width())
    width, height =p.get_width(),p.get_height()
    x=p.get_x()+width+0.00001
    y=p.get_y()+height/2
    ax.annotate(percent,(x,y),fontsize=12)
```

From the cancellation matrix, we see that Southwest Airline has the poorest performance based on this matrix while Delta Airline has the best performance.

```python
rank1 = INDEX1.loc[:,['AIRLINE','CANCELLATION_INDEX']]
rank2 = INDEX.loc[:,['AIRLINE','DELAY_INDEX']]

RANK=rank1.merge(rank2,on='AIRLINE')
RANK['RANK']=RANK['CANCELLATION_INDEX']+RANK['DELAY_INDEX']
RANK.sort_values(by='RANK',inplace=True, ascending=False)
RANK
```

```python
fig,ax=plt.subplots(figsize=(10,8))
plt.barh(RANK['AIRLINE'],RANK["RANK"])
plt.ylabel("Airline",fontsize=15)
plt.xlabel("RANK",fontsize=15)
#plt.xlim(0,0.30 )
plt.title('Performance Ranking For To Five Airlines In 2019',fontsize=18)
for p in ax.patches:
    percent ='({:,.2f})'.format(p.get_width())
    width, height =p.get_width(),p.get_height()
    x=p.get_x()+width+0.00001
    y=p.get_y()+height/2
    ax.annotate(percent,(x,y),fontsize=12)
```

The ranking based on both matrix indicate Southwest airline is the least performing while the United Airline is the best performing. Though it can be said that United is the least active thats why it could be at the top. Delta Airline been the second most active of the top five also has a very good record for 2019.
